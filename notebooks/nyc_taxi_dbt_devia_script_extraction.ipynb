```python
import requests          # télécharger les fichiers
import pandas as pd      # lire les fichiers Parquet
from google.cloud import bigquery  # communiquer avec BigQuery
from io import BytesIO
```

```python
base_url = "https://d37ci6vzurychx.cloudfront.net/trip-data/"
```

```python
client = bigquery.Client(project="nyc-taxi-dbt-devia")
```

```python
job_config = bigquery.LoadJobConfig(
    write_disposition="WRITE_APPEND",
    schema_update_options=[
        bigquery.SchemaUpdateOption.ALLOW_FIELD_ADDITION
    ]
)
```

```python
for year in [2024, 2025]:
  year_formate = str(year).zfill(4)
  for mois in range(1, 13):
    mois_formate = str(mois).zfill(2)
    url = base_url + "yellow_tripdata_" + year_formate + "-" + mois_formate + ".parquet"
    print(url)
    response = requests.get(url)
    df = pd.read_parquet(BytesIO(response.content))
    client.load_table_from_dataframe(df, "nyc-taxi-dbt-devia.raw.yellow_taxi_trips", job_config=job_config)
```
